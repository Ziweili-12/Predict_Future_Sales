## Create a dataframe for time seires model

```python
b.index = pd.DatetimeIndex(b.index)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
get_ipython().run_line_magic('matplotlib', 'inline')
plot_acf(b)
plt.show()
from statsmodels.tsa.stattools import adfuller
```

```python
D_data = b.diff().dropna()
D_data.columns = [u'dianliangchafen']
D_data.plot() 
plt.show()
plot_acf(D_data)    
plt.show()
plot_pacf(D_data)  
plt.show()
print(u'resultï¼š ', adfuller(D_data[u'dianliangchafen']))  
from statsmodels.stats.diagnostic import acorr_ljungbox
print(u'resultï¼š',acorr_ljungbox(D_data, lags= 1)) 
```

```python
from statsmodels.tsa.arima_model import ARIMA 
pmax = int(len(D_data) / 10)
qmax = int(len(D_data) / 10)
bic_matrix = []
for p in range(pmax +1):
    temp= []
    for q in range(qmax+1):
        try:
            temp.append(ARIMA(b, (p, 1, q)).fit().bic)
        except:
            temp.append(None)
        bic_matrix.append(temp)
        
bic_matrix = pd.DataFrame(bic_matrix)
bic_matrix=bic_matrix.astype('float64')
bic_matrix.info()
 

p,q = bic_matrix.stack().idxmin()
print(u'Minimum p and q of BICï¼š%s,%s' %(p,q))  
model = ARIMA(b, (p,1,q)).fit()
model.summary2()     
aa=model.forecast(1) 
```
